```python
#methods in classes and use of self keyword

class Player:
    score = 10  # Class attribute

    #Instance method
    def add_points(self,score):  # Method to add points (self use to refer the instance attribute)
        final_score =  self.score + score # Add the points to the current score (self. score refers to the instance attribute)
        return final_score
    #self is not a keyword, but a convention. You can use any other name instead of self, but it's not recommended.
    
player_1 = Player()
print(player_1.add_points(10))  # Output: 20
print(player_1.add_points(20))  # Output: 30
```

```python


class Car:
    wheels = 4  # Class attribute #differnce of class attribute and instance attribute is that class attribute is shared among all instances of the class, while instance attribute is unique to each instance.

    def drive_to(self, destination):
        return f"Driving to {destination} with {self.wheels} wheels."
    
    def show_wheel_count(self):
        return f"This car runs on {self.wheels} wheels."
    
car_1 = Car()
print(car_1.drive_to("New York"))  
print(car_1.show_wheel_count())    

#we are using constructors bcs 
```

```python
#constructor of class

class Car:
    wheels = 4  # Class attribute

    def __init__(self, brand, color):  # Constructor with parameters brand and color
        self.brand = brand      # Instance attribute
        self.color = color    # Instance attribute


car_1 = Car("Toyota", "Red")
print(car_1.brand)  # Output: Toyota
print(car_1.color)  # Output: Red
print(car_1.wheels)  # Output: 4

#In python there is no constructor overloading , only one constructor is allowed per class. If multiple constructors are defined, the last one will override the previous ones.

    
```

```python
class Car :
    wheels = 4  # Class attribute

    def __init__(self, brand, color="Black"):  # Constructor with default parameter color
        self.brand = brand      # Instance attribute
        self.color = color    # Instance attribute

    def drive_to(self, destination):#local variable destination and instance variables self.color and self.brand
        return f"The {self.color} {self.brand} is driving to {destination} "
    
    def show_wheel_count(self):#instance variable self.wheels
        return f"The {self.color} {self.brand} runs on {self.wheels} wheels."


car_1 = Car("BMW", "Black")
print(car_1.drive_to("Los Angeles"))
print(car_1.show_wheel_count())
```

```python
#calculate the area and circumference of a circle using class and methods

from math import pi

class Circle:
    
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return pi * (self.radius ** 2)

    def circumference(self):
        return 2 * pi * self.radius

    def display_info(self) :
        return f"\nradius: {self.radius} \narea: {self.area():.2f} \ncircumference: {self.circumference():.2f} " #2f formats the float to 2 decimal places  
    
circle_1 = Circle(5)
print(circle_1.area())               # Output: 78.53981633974483
print(circle_1.circumference())      # Output: 31.41592653589793
print(circle_1.display_info())       # Output: radius: 5 

```

```python
#inheritence in python
class Animal:

    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} speaks.")

class Dog(Animal):  # Dog class inherits from Animal class
    #pass #pass is used to indicate that the class has no additional attributes or methods

    def speak(self):#override the speak method of Animal class
        print(f"{self.name} barks.")

dog_1 = Dog("Ashie")
dog_1.speak()  

#constructor and methods in super class inherited to subclass
```

```python
class Animal:

    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} speaks.")


class Dog(Animal):  # Dog class inherits from Animal class

    def __init__(self, name, color):#constructor of Dog class with additional parameter color

        super().__init__(name)#call the constructor of the super class and initialize the name attribute of Animal class
        self.color = color#initialize the color attribute of Dog class


    def speak(self):
        print(f"{self.color} {self.name} barks")


dog_1 = Dog("Ashie", "Brown")
dog_1.speak()   # Output: Brown Ashie barks
```
