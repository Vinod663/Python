```python
#Print the numbers in descending order
num1=float(input("Enter first number: "))#by default input() takes string input
num2=float(input("Enter second number: "))
num3=float(input("Enter third number: "))


if num1>=num2 and num1>=num3:
    if num2>=num3:#Nested if-else statement--> if-else inside if or else block
        print(num1,+num2,num3)
    else:
        print(num1,num3,num2)

elif num2>=num1 and num2>=num3:
    if num1>=num3:
        print(num2,num1,num3)
    else:
        print(num2,num3,num1)

else:
    if num1>=num2:
        print(num3,num1,num2)
    else:
        print(num3,num2,num1)


```

```python
#Structural Pattern Matching---> match-case statement
#Syntax:
# match variable:
#     case pattern1:
#         statement1
#     case pattern2:
#         statement2
#     case _:
#         statement3

x=9

match x:
    case 1:
        print("One")
    case 2:
        print("Two")
    case 3:
        print("Three")
    case _:
        print("Other than 1,2,3")
```

```python
#Print day of the week is a weekday or weekend
day = input("Enter the day of the week: ")

match day.lower(): #.lower() converts the input string to lowercase
    case "monday" | "tuesday" | "wednesday" | "thursday" | "friday": # | is used for OR operation in structural pattern matching
        print("It's a weekday.")
    case "saturday" | "sunday":
        print("It's a weekend.")
    case _:
        print("Invalid day.")
```

```python
#Pattern matching 
#This checks the structure of the list and matches accordingly
numbers=[4,3,7]

match numbers:
    case [x,y,z]: #matches if numbers has exactly 3 elements
        print(f"List with three elements: {x}, {y}, {z}")
    case [x,y]: #matches if numbers has exactly 2 elements
        print(f"List with two elements: {x}, {y}")
    case _:
        print("List with more than three elements")
```

```python
#Take list of numbers and calculate hypotenuse or area of triangle based on list number of elements

numbers=[3,4,5]#for Area of triangle
#numbers=[3,4]#for Hypotenuse


match numbers:
    case [a, b]: # if list has 2 elements
        hypotenuse = (a**2 + b**2)**0.5
        print(f"Hypotenuse is: ", hypotenuse)

    case [a, b, c]: # if list has 3 elements
        s = (a + b + c) / 2  # semi-perimeter --> half of perimeter
        area = (s * (s - a) * (s - b) * (s - c))**0.5  # Heron's formula for area of triangle
        print(f"Area of triangle with sides is: ", area)

    case _:
        print("List should have either 2 or 3 elements")
```

```python
#Filter out Odd numbers from a list and print the new list
nums=[10,9,8,5,12,18]


for odds in nums:
    if odds % 2 != 0: #Checking for odd numbers
        nums.remove(odds) #Removing odd numbers from the list

print("Even numbers are: ", nums)
```

```python
#Calculate factorial of a number
num=int(input("Enter a number to find factorial: "))
factorial=1

if num<0:
    print("Factorial does not exist for negative numbers")
elif num==0:
    print("Factorial of 0 is 1")
else:
    for i in range(1,num+1): 
        factorial=factorial*i # factorial mean --> if num is 5 , 1*2*3*4*5 = 120
    print("Factorial of", num, "is", factorial)
```

```python
#Break statement
numbers=[2,3,9,4,7,6,1]

for num in numbers:
    if num>5:
        print("Number greater than 5 found:", num)
        break
    print(num)
```

```python
#Take 10 numbers as input from user and stop if user enters a zero
#Also calculate the sum of entered numbers

total=0
for i in range(10):
    number=float(input("Enter a number (0 to stop): "))
    if number==0:
        print("Zero entered, stopping input.")
        break
    total+=number
print("Total sum of entered numbers is:", total)
```
