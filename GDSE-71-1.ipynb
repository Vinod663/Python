```python
#Insert elements to Mid--- using insert() method
my_list = [1, 2, 3, 4, 5]   
print(my_list)  # Print the original list
my_list.insert(2, 10)  # Insert 10 at index 2
print(my_list)  # Print the updated list
```

```python
#Insert elements using append() method
my_list = [1, 2, 3, 4, 5]   
print(my_list)  # Print the original list
my_list.append(6)  # Append 6 to the end of the list
print(my_list)  # Print the updated list
```

```python
#Remove element using remove() method
my_list = ["Cat", "Dog", "Elephant", "Tiger", "Lion"]
print(my_list)  # Print the original list
my_list.remove("Elephant")  # Remove "Elephant" from the list
print(my_list)  # Print the updated list
```

```python
# Same element multiple times
my_list = ["Cat", "Dog", "Elephant", "Tiger", "Lion","Dog"] 
print(my_list)  # Print the original list
my_list.remove("Dog")  # Remove "Dog" from the list (removes the first occurrence)
print(my_list)  # Print the updated list
```

```python
#Remove and return the last element using pop() method
my_list = ["Cat", "Dog", "Elephant", "Tiger", "Lion","Dog"] 
print(my_list)  # Print the original list
removed_element = my_list.pop(2)  # Remove and return the element at index 2
print("Removed Element:", removed_element)  # Print the removed element
print(my_list)  # Print the updated list
```

```python
#Remove elemet using del keyword
my_list = ["Cat", "Dog", "Elephant", "Tiger", "Lion","Dog"]
print(my_list)  # Print the original list
del my_list[1]  # Delete the element at index 1 ("Dog")
print(my_list)  # Print the updated list
```

```python
#Remove List using del keyword
my_list = ["Cat", "Dog", "Elephant", "Tiger", "Lion","Dog"]
print(my_list)  # Print the original list
del my_list  # Delete the entire list
#print(my_list)  # This will raise an error because my_list is deleted
```

```python
#Remove elements using clear() method
my_list = ["Cat", "Dog", "Elephant", "Tiger", "Lion","Dog"]
print(my_list)  # Print the original list
print(len(my_list))  # Print the length of the original list

my_list.clear()  # Clear all elements from the list
print(my_list)  # Print the empty list
print(len(my_list))  # Print the length of the empty list
```

```python
#Sort Descending order
my_list = ["Cat", "Dog", "Elephant", "Tiger", "Lion","Dog"]
print(my_list)  # Print the original list
my_list.sort(reverse=True)  # Sort the list in descending order
print(my_list)  # Print the sorted list
```

```python
#Reverse the list using reverse() method
my_list = ["Cat", "Dog", "Elephant", "Tiger", "Lion","Dog"]
print(my_list)  # Print the original list
my_list.reverse()  # Reverse the order of the list
print(my_list)  # Print the reversed list
```

```python
#Copy the list using copy() method
my_list = ["Cat", "Dog", "Elephant", "Tiger", "Lion","Dog"]
print(my_list)  # Print the original list

new_list = my_list.copy()  # Create a copy of the list
print(new_list)  # Print the copied list

print("my_list:",id(my_list))  # Print the memory address of the original list
print("new_list:",id(new_list))  # Print the memory address of the copied list

my_list[0] = "Monkey"  # Modify the first element of the original list
print("After modification:")
print("my_list:",my_list)  # Print the original list after modification
print("new_list:",new_list)  # Print the new_list after modification
```

```python
#Assigning a list to another variable creates a reference to the same list, not a copy
my_list=["Cat", "Dog", "Elephant", "Tiger", "Lion","Dog"]

new_list=my_list # This creates a reference to the same list, not a copy
print("my_list:",id(my_list))  # Print the memory address of the original list
print("new_list:",id(new_list))  # Print the memory address of the new_list

new_list[0]="Monkey"
print("my_list:",my_list)  # Print the original list after modification
print("new_list:",new_list)  # Print the new_list after modification
```

```python
#Joining two lists using + operator
list1 = [1, 2, 3]
list2 = [4, 5, 6]
combined_list = list1 + list2  # Join the two lists
print(combined_list)  # Print the combined list
```

```python
#Extend a list using extend() method
list1 = [1, 2, 3]
list2 = [4, 5, 6]
list1.extend(list2)  # Extend list1 by appending elements from list2
print(list1)  # Print the extended list1
print(list2)  # Print list2 to show it remains unchanged
```
