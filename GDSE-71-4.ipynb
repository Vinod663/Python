```python
#Functional Programming
#return sum of two numbers

def sum_numbers(a, b): #Parameters a and b
    return a + b        #return sum of a and b

#Call the function and print the result
print(sum_numbers(5, 10))  #Output: 15  (#5,10 defined as arguments)
```

```python
#Calculate the sum of squres of positive integers from 1 to n

def sum_of_squares(n): #Parameter n
    sum_squares = 0
    if n < 1: #check for positive integer
        return ("Input must be a positive integer.") #when return met function exits, so no need for else
    
    for i in range(1, n + 1):  #Loop from 1 to n
        sum_squares += i ** 2  #Add square of i to sum_squares
    return sum_squares

print(sum_of_squares(5))  


```

```python
#return the sum of of two list of numbers as new list containg the sum of corresponding elements

def sum_lists(list1, list2): #Parameters list1 and list2
    if len(list1) != len(list2): #Check if lists are of equal length
        return "Error: Lists are not of equal length." #Return error message if not equal
    
    summed_list = []  #Initialize empty list to store sums
    
    for i in range(len(list1)):  #Loop through indices of the lists
        summed_list.append(list1[i] + list2[i]) # append use for adding elements to list to end of list
    return summed_list

print(sum_lists([1, 2, 3], [4, 5, 6]))  #Output: [5, 7, 9]
```

```python
#Default arguments in functions


def add(a, b=10): #Parameter b has default value of 10
    return a + b

print(add(3,4))        #Output: 7 (b is provided as 4)
print(add(3))          #Output: 13 (b uses default value of 10)
print(add(3,4,5))    #Output: TypeError (too many positional arguments)
```

```python
#Rule of default arguments: All default arguments must be at the end of the parameter list.

def multiply(a, b=2, c=3): #b and c have default values
    return a * b * c

print(multiply(4))        #Output: 24 (uses default b=2, c=3)

# def multiply_invalid(a=1, b, c=10): #This will raise a SyntaxError
#     return a * b * c

# print(multiply_invalid(5, 2))  #Output: SyntaxError: non-default argument follows default argument
```

```python
#Calculate the total cost and return the total cost after applying discount 

def calculate_total(price, quantity=1, discount=0): #quantity and discount have default values
    total = price * quantity  #Calculate total price
    total_after_discount = total - (total * discount / 100)  #Apply discount
    return total_after_discount

print(calculate_total(100, 2, 10))  #Output: 180.0 (2 items with 10% discount)
print(calculate_total(50))           #Output: 50.0 (1 item with no discount)
print(calculate_total(200, discount=20))  #Output: 160.0 (1 item with 20% discount)
```

```python
#Arbitrary Arguments
#1.Arbitrary Positional Arguments(*args)

def arbitrary_positional_args(*args): #*args allows for arbitrary number of positional arguments
   print (args)
   print (type(args))

arbitrary_positional_args(1, 2, 3, 4)  #Output: (1, 2, 3, 4) <class 'tuple'>
arbitrary_positional_args('a', 'b', 'c')  #Output: ('a', 'b', 'c') <class 'tuple'>
arbitrary_positional_args()  #Output: () <class 'tuple'>
```

```python
#return the sum of any numeric number of arguments
def sum_numbers(*numbers):
    #check if all arguments are numeric
    sum_total = 0

    for num in numbers:
        sum_total += num

    return sum_total

print(sum_numbers(1, 2, 3))  #Output: 6
print(sum_numbers(10, 20, 30, 40, 50))  #Output: 150
# print(sum_numbers("a","b","c"))  #Output: Error: All arguments must be numeric.
```

```python
#Calculate the average of any number of numeric positive numbers

def average_positive(*numbers):
    positive_sum = 0
    positive_count=0
    average_positive=0

    for num in numbers: #Iterate through all provided numbers
        if num > 0: #Check if the number is positive
            positive_sum += num
            positive_count += 1

    if(positive_count==0):#if no positive numbers were found
        return "No positive numbers provided."
    
    average_positive = positive_sum / positive_count 
    return average_positive

print(average_positive(10, -5, 15, 20)) #Output 15.0
print(average_positive(-1, -2, -3, 0))      #Output: No positive numbers provided.
print(average_positive(5, 10, 15, 20, 25)) #Output: 15.0


            
```

```python
#Arbitraryb Keyword Arguments (**kwargs)

def arbitrary_keyword_args(**kwargs): #**kwargs allows for arbitrary number of keyword arguments
    print(kwargs)
    print(type(kwargs))

arbitrary_keyword_args(name="Alice", age=30, city="New York")
```

```python
#Sepearate the keys and values from the arbitrary keyword arguments

def separate_keys_values(**kwargs):
    keys = []
    values = []

    for value in kwargs.values():
        values.append(value)
        #print("Value:", value)

    for key in kwargs.keys():
        keys.append(key)
        #print("Key:", key)

    print("Keys:", keys)
    print("Values:", values)

separate_keys_values(name="Bob", age=25, country="USA")
```

```python
#print key-value pairs from arbitrary keyword arguments using .items()

def print_key_value_pairs(**kwargs):
    for key, value in kwargs.items():
        print(key + ":", value)

print_key_value_pairs(fruit="Apple", color="Red", quantity=5)
```

```python
#Calculate and print the average score of all subjects from arbitrary keyword arguments
def average_score(**subjects):
    
    if not subjects:  #Check if no subjects were provided
        return "No Subjects provided."

    total_score = 0
    
    for score in subjects.values():
        total_score += score

    average = total_score / len(subjects)
    return average


print(average_score(Math=90, Science=85, English=88))  
print(average_score())  
```

```python
#Check Absolute Value of Numbers without built-in abs() function
def absolute_value(num):
    if num < 0:
        return -num
    else:
        return num
    
print(absolute_value(-10))  #Output: 10
print(absolute_value(5))    #Output: 5
```

```python
#Absolute with built-in abs() function

def absolute_value_builtin(num):
    return abs(num)

print(absolute_value_builtin(-20))  #Output: 20
print(absolute_value_builtin(15))    #Output: 15
```

```python
#map function to square each number in a list

def find_square(num):
    return num ** 2

my_list =[2,6,3,4]

result = map(find_square, my_list) #map applies find_square to each element in my_list
result_list = list(result)  #Convert map object to list


print(result)  #Output: <map object at ...>
print (result_list)  #Output: [4, 36, 9, 16]

print(type(result))  #Output: <class 'map'>
print(type(result_list))  #Output: <class 'list'>
```

```python
#Convert Farenheit to Celsius using map function and accepts a list of temperatures in Farenheit, return a new list of temperatures in Celsius rounded to two decimal places

def fahrenheit_to_celsius(farenheit_list): 

    def convert_to_celsius(farenheit): 
        celsius = (farenheit - 32) * 5.0/9.0

        return round(celsius, 2)  

    celsius_list = list(map(convert_to_celsius, farenheit_list)) 
    return celsius_list

print (fahrenheit_to_celsius([32, 68, 100, 212]))  #Output: [0.0, 20.0, 37.78, 100.0
```

```python
# add corresponding elements of two lists using map function

def add_lists(list1, list2):
    def add_elements(a, b): #Function to add two elements
        return a + b

    #we can pass multiple iterables to map function     
    summed_list = list(map(add_elements, list1, list2)) #map applies add_elements to corresponding elements of list1 and list2
    return summed_list

print(add_lists([1, 2, 3], [4, 5, 6]))  #Output: [5, 7, 9]
```

```python
#Filter evens from a list using filter function
def is_even(num):
    return num % 2 == 0  #Return True if num is even

my_list = [9,5,8,12,13]

even_numbers = list(filter(is_even, my_list))  #filter applies is_even to each element in my_list
print("even_numbers: ",even_numbers)  #Output: [8, 12]
```

```python
#Extract numbers from list that divisible by 3 but not by 5 using filter function

def divisible_by_3_not_5(num):
    return num % 3 == 0 and num % 5 != 0  

numbers=[12,15,18,20,25,30,35]
filtered_numbers = list(filter(divisible_by_3_not_5, numbers))  

print("filtered_numbers: ",filtered_numbers)  #Output: [12, 18]
```
