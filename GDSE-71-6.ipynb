```python
#public class
class Dog :

    def __init__(self, name):

        self.name = name

dog_1 = Dog("Buddy") 
print(dog_1.name) #Can access the name attribute directly
    

        
```

```python
#protected class
class Dog :
    def __init__(self, name):
        self._name = name  # Protected attribute (use a single underscore)

dog_1 = Dog("Buddy")
print(dog_1._name)  # Can still access the protected attribute directly, but usually we don't do this outside the class, we use it within subclasses or the class itself.
```

```python
#protected with subclass
class Dog :
    def __init__(self, name):
        self._name = name  # Protected attribute (use a single underscore)

class Puppy(Dog):
    def get_name(self):
        print(self._name)  # Accessing the protected attribute from the subclass

puppy_1 = Puppy("Max")
puppy_1.get_name()  # This will print "Max" #this is the recommended way to access protected attributes within subclasses

print(puppy_1._name)  # Although we can access it directly, it's not recommended outside the class or subclass #
```

```python
#Parivate class

class Dog :
    def __init__(self, name):
        self.__name = name  # Private attribute (use double underscores)

dog_1 = Dog("Buddy")
#print(dog_1.__name)  # This will raise an AttributeError because __name is private

#private attributes only accessible within the class itself , not even in subclasses.

# we use getters and setters to access private attributes
```

```python
#Q-1
class StudentRecord:
    def __init__(self):
        self.__grades = [] # Private attribute to store grades(list of grades 0-100)
    
    def add_grade(self, mark):
        #accept the grade and add to the grades list
        if self.__validate_grade(mark):
            self.__grades.append(mark)
        else:
            print("Invalid mark. Please enter a mark between 0 and 100.")
    

    def __validate_grade(self, mark): #private method to validate the grade (cannot be accessed outside the class)
        #validate the grade is between 0 and 100 and check if it is integer
        if isinstance(mark, int) and 0 <= mark <= 100:
            return True
        else:
            return False


    def get_average(self): #public method to get average
        return self.__calculate_average()
        

    def __calculate_average(self): #private method to calculate average (cannot be accessed outside the class)
        if len(self.__grades) == 0:
            return 0
        return sum(self.__grades) / len(self.__grades)
    
    
    

# Example usage:
student = StudentRecord()
student.add_grade(85)
student.add_grade(90)
student.add_grade(78)
average = student.get_average()
print(f"Average grade: {average}")  # Output: Average grade: 84.33333333333333
        
```
