```python
#File Handeling (text file)

file_1 = open("my_file.txt", "r")  # Open a file in read mode #Default mode is read mode "r"(if not specified the mode it will be read mode)

content = file_1.read()  # Read the content of the file
print(content)  # Print the content to the console
file_1.close()  # Close the file
```

```python
#With statement

with open("my_file.txt", "r") as file: # Open the file using 'with' statement
    content = file.read()
    print(content)
    # No need to explicitly close the file, it's done automatically when the with block is exited.
```

```python
#Line by line reading
with open("my_file.txt", "r") as file:
    line_1 = file.readline()  # Read the first line
    print("Line 1:", line_1)

    line_2 = file.readline()  # Read the second line
    print("Line 2:", line_2)

    #Strip newline characters
    line_3 = file.readline().strip()  # Read the third line and strip newline characters (remove the /n at the end)
    print("Line 3:", line_3)

    line_4 = file.readline()
    print("Line 4:", line_4)
```

```python
#Use readline without multiple calls
with open("my_file.txt", "r") as file:
    while True:
        line = file.readline().strip()
        if not line:  # If line is empty, end of file is reached
            break
        print(line)
        
```

```python
#readlines
with open("my_file.txt", "r") as file:
    lines = file.readlines()  # Read all lines into a list
    print(lines)  # Print the list of lines
```

```python
#write mode
with open("my_file.txt", "w") as file: #if file not exists it will create a new file otherwise it will overwrite the existing file
    file.write("Hello, World!\n")  # Write a line to the file
    file.write("This is a new line.\n")  # Write a new line to the file
    file.write("Another line added.\n")  # Write another line
```

```python
#New file creation and writing
with open("new_file.txt", "w") as new_file:
    new_file.write("This is a newly created file.\n")
    new_file.write("It contains some sample text.\n")
```

```python
#wrirelines (multiple lines at a time)
with open("new_file.txt", "w") as file:
    lines = ["First line\n", "Second line\n", "Third line\n"]
    file.writelines(lines)
```

```python
#append mode(to add new content without overwriting existing content)
with open("my_file.txt", "a") as file:
    file.write("This line is appended to the file.\n")
    file.write("Appending another line.\n")
```

```python
#Covert JSON file data to Python Dictionary(deserialization) 
#deserialization/decode ---->  converting JSON data into a Python object, such as a dictionary or a list.
#Serialization/encode ------> converting a Python object into a JSON string.

import json

with open("my_json.json", "r") as json_file:
    data = json.load(json_file)  # Load () --> Deserialize JSON data to Python dictionary
    print(data)  # Print the dictionary
    print(type(data))  # Print the type of the data (should be <class 'dict'>)
```

```python
#Convert Python Dictionary to JSON file (serialization)
import json

data ={
    "programe": "GDSE",
    "batch": 71,
    "semester": 3,
}

with open("new_json.json", "w") as json_file:
    json.dump(data, json_file, indent=4)  # Dump() --> Serialize Python dictionary to JSON file -->dump(data, file_object)
    #indent parameter is used to format the JSON output with indentation for better readability.
```

```python
#File Handling(CSV Files)
import csv

#csv.reader()---> Reads data from a CSV file and returns it as a list of lists, where each inner list represents a row in the CSV file.

with open("person_data.csv", "r") as csv_file:
    csv_reader = csv.reader(csv_file)
    print(csv_reader)  # Print the reader object
    print(type(csv_reader))  # Print the type of the reader object

    for row in csv_reader: 
        print(row)  # Print each row as a list
```

```python
#csv.dectReader() ---> Reads data from a CSV file and returns it as a list of dictionaries, where each dictionary represents a row in the CSV file with column headers as keys.
import csv

with open("person_data.csv", "r") as csv_file:
    csv_dict_reader = csv.DictReader(csv_file)
    print(csv_dict_reader)  # Print the DictReader object
    print(type(csv_dict_reader))  # Print the type of the DictReader object

    for row in csv_dict_reader:
        print(row)  # Print each row as a dictionary
    
    
```

```python
import csv

csv_file_path = "person_data.csv"

#data to be written to the CSV file (list of dictionaries)
data = [
    {"Name": "Alice", "Age": 28, "City": "New York"},
    {"Name": "Bob", "Age": 34, "City": "Los Angeles"},
    {"Name": "Charlie", "Age": 25, "City": "Chicago"},  
]

#deifine the CSV field names(column-headers names )
fieldnames = ["Name", "Age", "City"]

with open(csv_file_path, "w", newline="") as csv_file: #newline="" to prevent extra blank lines on some platforms (Use to avoid blank lines in Windows)
    csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames) # Create a DictWriter object with the specified field names

    csv_writer.writeheader()  # Write the header row to the CSV file
    csv_writer.writerows(data)  # Write multiple rows of data to the CSV file


print("Data has been written to", csv_file_path)

    
```

```python
#read text file and writes only even numbered line to a new text file(first number as line 1)(without using enumerate)

with open("input.txt", "r", newline='') as input_file:
    with open("even_lines.txt", "w", newline='') as output_file:
        line_number = 1  # Initialize line number counter as 1  for default
        for line in input_file:
            if line_number % 2 == 0:  # Check if the line number is even
                output_file.write(line)  # Write the even-numbered line to the output file
            line_number += 1  # Increment the line number counter
        print("Even-numbered lines have been written to even_lines.txt\n")


#print even numbererd list called even_lines.txt
with open("even_lines.txt", "r") as file:
    for line in file:
        print(line.strip())  # Print each line after stripping newline characters




        
```
