```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

print("Array:", arr) #[1 2 3 4 5] no commas
print("Type", type(arr))
```

```python
# Addition of two arrays

arr_1 = np.array([10, 20, 30, 40, 50])

arr_2 = np.array([1, 2, 3, 4, 5])

arr_3 = arr_1 + arr_2
print("Sum of two arrays:", arr_3) #[11 22 33 44 55]
```

```python
# Subtraction of two arrays

arr_1 = np.array([10, 20, 30, 40, 50])
arr_2 = np.array([1, 2, 3, 4, 5])

arr_3 = arr_1 - arr_2
print("Difference of two arrays:", arr_3) #[ 9 18 27 36 45]
```

```python
# Multiplication of two arrays

arr_1 = np.array([10, 20, 30, 40, 50])
arr_2 = np.array([1, 2, 3, 4, 5])

arr_3 = arr_1 * arr_2
print("Product of two arrays:", arr_3) #[ 10  40  90 160 250]
```

```python
# Division of two arrays

arr_1 = np.array([10, 20, 20, 40, 50])
arr_2 = np.array([2,7,3,8,5])

arr_3 = arr_1 / arr_2
print("Division of two arrays:", arr_3) #[5. 2.85714286 6.66666667 5. 10.]
```

```python
#Sum of all elements in an array\

arr = np.array([1, 2, 3, 4, 5])
total = np.sum(arr)

print("Sum of all elements in the array:", total) #15
```

```python
#mean of array elements

arr = np.array([1, 2, 3, 4, 5])
mean_value = np.mean(arr)

print("Mean of array elements:", mean_value) #3.0 (1+2+3+4+5)/5 = 15/5 =3.0
```

```python
#maximum value in an array

arr = np.array([1, 2, 3, 4, 5])
max_value = np.max(arr)

print("Maximum value in the array:", max_value) #5
```

```python
#minimum value in an array

arr = np.array([1, 2, 3, 4, 5])
min_value = np.min(arr)

print("Minimum value in the array:", min_value) #1
```

```python
#Square root of each element in the array

arr = np.array([1, 4, 9, 16, 25])
sqrt_arr = np.sqrt(arr)

print("Square root of each element in the array:", sqrt_arr) #[1. 2. 3. 4. 5.]
```

```python
#Condition 

arr = np.array([10, 20, 15, 18, 14, 26])

condition = arr > 15

print("Condition (elements greater than 15):", condition) #[False  True False  True False  True]

filtered_arr = arr[condition] #filtering elements based on condition
#filtered_arr = arr[arr > 15] #directly filtering elements based on condition
print("Filtered array (elements greater than 15):", filtered_arr) #[20 18 26]
```

```python
#(i)greater than 10 or less than 5
#(ii) not equal to 15
arr = np.array([
    [12, 5, 7, 20],
    [3, 15, 8, 2,],
    [10, 25, 6, 18],
    [4, 11, 30, 9]
])

filtered_arr = arr[((arr>10) | (arr<5)) & (arr != 15) ] #filtering even numbers

print("Filtered array (elements greater than 10 or less than 5 and not equal to 15):", filtered_arr) #[12 20 3 2 25 18 4 11 30]
```

```python
data ={
    "age" : [25, 30, 22, 28, 32],
    "Salary" : [50000, 60000, 55000, 62000, 58000],
}

df = pd.DataFrame(data)

#Sum
print(df.sum())  #use to get the sum of each numerical column

#Mean
print("\n",df.mean()) #use to get the mean of each numerical column

#describe
print("\n",df.describe()) #use to get the statistical summary of the DataFrame

#std --> standard deviation (in simple terms, it measures how spread out the numbers are in a dataset) ex-ample: if the ages are 22, 23, 24, 25, 26, the standard deviation is low because the ages are close to the mean (24). But if the ages are 10, 20, 30, 40, 50, the standard deviation is high because the ages are spread out over a wider range.
#-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#how to get std step by step:
#1. Calculate the mean (average) of the dataset.
#2. Subtract the mean from each data point and square the result.
#3. Calculate the average of these squared differences.
#4. Take the square root of that average. This gives you the standard deviation.

#example:
#data: 10, 12, 14, 16, 18
#mean = (10 + 12 + 14 + 16 + 18) / 5 = 14
#std = sqrt(((10-14)² + (12-14)² + (14-14)² + (16-14)² + (18-14)²) / 5) = sqrt((16 + 4 + 0 + 4 + 16) / 5) = sqrt(40 / 5) = sqrt(8) ≈ 2.83
#-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

#order according to ascending order
#22, 25, 28, 30, 32

#25% -> 25th percentile (the value below which 25% of the data falls(First quartile/Q1))
#How to calculate 25th percentile step by step:
#1. Order the data from smallest to largest.
#2. Calculate the rank using the formula: Rank = (P/100) * (N + 1), where P is the desired percentile (25 in this case) and N is the number of data points.
#3. If the rank is an integer, the 25th percentile is the value at that position. If the rank is not an integer, interpolate between the two closest ranks to find the 25th percentile.
#Example:
#data: 22, 25, 28, 30, 32
#N = 5 (number of data points)
#Rank = (25/100) * (5 + 1) = 1.5
#Since the rank is 1.5, we take the average of the values at positions 1 and 2:
#25th percentile = (22 + 25) / 2 = 23.5
#-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------


#50% -> median (the middle value of the dataset(median/Q2))

#---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#75% -> 75th percentile (the value below which 75% of the data falls(Third quartile/Q3))
#How to calculate 75th percentile step by step:
#Example:
#data: 22, 25, 28, 30, 32
#N = 5 (number of data points)  
#Rank = (75/100) * (5 + 1) = 4.5
#Since the rank is 4.5, we take the average of the values at positions 4 and 5:
#75th percentile = (30 + 32) / 2 = 31.0
#but 30 is the 75th percentile in pandas because pandas uses a different method for calculating percentiles which may lead to slightly different results compared to manual calculations.
#-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
```

```python


df = pd.DataFrame({
    "Product" : ["Laptop", "Mouse", "Keyboard", "Monitor", "Headset"],
    "UnitSold" :[12, 50, 34, 20, 40],
    "Price" : [120000, 1500, 3500, 25000, 6000],
    "Discount" : [0.10, 0.05, 0.08, 0.12, 0.07]
})

#Final price after discount 
discounted_prices =df['Price'] * (1 - df['Discount'])
print("Final prices after discount:\n", discounted_prices)

#Total revenue (UnitSold * Final Price)
total_revenue = df['UnitSold'] * discounted_prices
print("\nTotal revenue of each product:\n", total_revenue)

#Add as new columns to the DataFrame
df['FinalPrice'] = discounted_prices
df['TotalRevenue'] = total_revenue

print("\nDataFrame with Final Price and Total Revenue:\n", df)


#products that revenue> 5000 00
high_revenue_products = df[df['TotalRevenue'] > 500000]
print("\nProducts with Total Revenue greater than 500,000:\n", high_revenue_products)

#Total revenue from all products
total_revenue_all_products = df['TotalRevenue'].sum()
print("\nTotal revenue from all products:", total_revenue_all_products)
```

```python
#remove rows and columns from DataFrame using .drop()
#axis=0 --> rows
#axis=1 --> columns

#Syntax:
#df.drop(labels, axis=0 or 1, inplace=False)(inplace default is False)
#inplace=True --> modifies the original DataFrame
#inplace=False --> returns a new DataFrame without modifying the original DataFrame 

#Remove Discount column
df = pd.DataFrame({
    "Product" : ["Laptop", "Mouse", "Keyboard", "Monitor", "Headset"],
    "UnitSold" :[12, 50, 34, 20, 40],
    "Price" : [120000, 1500, 3500, 25000, 6000],
    "Discount" : [0.10, 0.05, 0.08, 0.12, 0.07]
})



#Remove Discount column without inplace
new_df = df.drop('Discount', axis=1) #removing 'Discount' column
print("DataFrame after removing 'Discount' column (without inplace):\n", new_df)

```

```python
#Remove row that prodect is Monitor

df = pd.DataFrame({
    "Product" : ["Laptop", "Mouse", "Keyboard", "Monitor", "Headset"],
    "UnitSold" :[12, 50, 34, 20, 40],
    "Price" : [120000, 1500, 3500, 25000, 6000],
    "Discount" : [0.10, 0.05, 0.08, 0.12, 0.07]
})

#Remove row that prodect is Monitor without inplace
new_df = df.drop(df[df['Product'] == 'Monitor'].index, axis=0) #removing row where Product is 'Monitor'
print("DataFrame after removing row where Product is 'Monitor' (without inplace):\n", new_df)
```
